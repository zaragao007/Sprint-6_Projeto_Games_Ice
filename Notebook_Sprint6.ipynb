Explicação das mudanças

YEAR OF RELEASE

Antes era número com ponto decimal (1985.0). Converti para inteiro (Int64), pois ano é sempre número inteiro.

CRITIC SCORE

Também estava como número decimal, mas são notas inteiras (0–100). Converti para inteiro (Int64).

USER SCORE

Originalmente era texto (string). Converti para float (número decimal), porque a nota dos usuários pode ser 8.3, 7.5 etc. Os valores inválidos foram transformados em NaN (vazio), assim evitamos erro de conversão.

```python
print("\nValores nulos por coluna:\n")
print(df_games.isnull().sum())

df_games['user_score'] = df_games['user_score'].replace('TBD', pd.NA) #substitui valores 'TBD' em user_score por NaN

df_games['rating'] = df_games['rating'].fillna('Unknown')
```

Primeiro passo foi verificar onde existem valores ausentes, para então decidir se vamos preencher ou deixar em branco.

Para colunas numéricas como critic_score e user_score, preferi deixar em branco (NaN), porque inventar um valor pode distorcer a análise.

Para colunas categóricas como rating, escolhi preencher com "Unknown" para indicar que não sabemos a classificação.

Jogos antigos podem não ter classificação por idade.

Dados podem ter sido perdidos na coleta ou não foram informados pelas fontes originais.

Tratamento para TBD

CLassificação TBD significa que a informação ainda não foi definida. Por isso, trocamos TBD por NaN, para ser tratado como valor ausente e não causar erro em cálculos.

```python
#top 5 plataformas por vendas totais
platform_totals = df_games.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
top_platforms = platform_totals.head(5).index.tolist()  # vira uma lista simples

#distribuição anual (somente Top 5)
dados = df_games[(df_games['platform'].isin(top_platforms)) & (df_games['year_of_release'].notna())]
annual = dados.groupby(['year_of_release', 'platform'])['total_sales'].sum().reset_index()

#tabela ano x plataforma (cada coluna = uma plataforma)
pivot = annual.pivot(index='year_of_release', columns='platform', values='total_sales').fillna(0).sort_index()
print("Distribuição anual:\n")
print(pivot.head(20))
```

```python
linhas = []
for p in top_platforms:
    #se a plataforma não estiver nas colunas do pivot, cria valores vazios
    if p not in pivot.columns:
        primeiro = pico = ultimo = subir = cair = None
    else:
        s = pivot[p]  #série anual dessa plataforma
        anos_ativos = s[s > 0].index.tolist()  #anos com venda > 0

        if len(anos_ativos) == 0:
            primeiro = pico = ultimo = subir = cair = None
        else:
            primeiro = int(anos_ativos[0])       #primeiro ano com vendas
            ultimo   = int(anos_ativos[-1])      #último ano com vendas
            pico     = int(s.idxmax())           #ano do maior valor
            subir    = pico - primeiro           #anos até o pico
            cair     = ultimo - pico             #anos do pico até sumir

    linhas.append([p, primeiro, pico, ultimo, subir, cair])

ciclo = pd.DataFrame(linhas, columns=['platform','first_year','peak_year','last_year','rise_years_to_peak','decline_years_after_peak'])

print("\nCiclo de vida (Top 5):\n\n")
print(ciclo)
```

```python
print('Tempo médio até o pico')
print(ciclo['rise_years_to_peak'].mean())
```

```python
print('Tempo médio do pico até desaparecer')
print(ciclo['decline_years_after_peak'].mean())
```

Em resumo: vemos ondas claras por geração — cada plataforma cresce, atinge um pico e depois entra em declínio.

Entre as Top 5, PS2 e DS aparecem com 0 em 2016. X360, PS3 e Wii ainda mostram valores residuais em 2016, porém menores que nos anos de pico.
Sabemos que os daddos de 2016 estão incompletos, então valores baixos ou zerados devem ser lidos com cautela.

Depois do pico, a queda costuma ser mais curta do que a subida total do ciclo.

```python
for yr in [2012, 2015]:
    if yr not in pivot.index:
        pivot.loc[yr] = 0
pivot = pivot.sort_index()

trend_diff = (pivot.loc[2015] - pivot.loc[2012]).sort_values(ascending=False)

def classifica(valor):
    if valor > 0:
        return "crescendo"
    elif valor < 0:
        return "caindo"
    else:
        return "estável"

trend_label = trend_diff.apply(classifica)

totals = leaders.rename("total_sales_2012_2015")
diff   = trend_diff.rename("dif_2015_vs_2012")
trend  = trend_label.rename("tendência")

summary = pd.concat([totals, diff, trend], axis=1)
summary = summary.sort_values("total_sales_2012_2015", ascending=False)

print("\nResumo por plataforma (2012–2015):")
print(summary.head(10))

candidates = summary[summary["tendência"].isin(["crescendo", "estável"])].head(5).index.tolist()
print("\nPlataformas potencialmente lucrativas (base 2012–2015):", candidates)
```

```python
dados = df_games[['platform', 'total_sales']].dropna()

medianas = dados.groupby('platform')['total_sales'].median().sort_values(ascending=False)
ordem = medianas.index.tolist()

caixas = []
for plataforma in ordem:
    serie = dados.loc[dados['platform'] == plataforma, 'total_sales']
    valores = serie.values
    caixas.append(valores)

from matplotlib.ticker import MultipleLocator

plt.figure(figsize=(14, 8))
plt.boxplot(caixas, tick_labels=ordem, showfliers=True)
plt.xticks(rotation=45)
plt.ylabel("Vendas globais (total_sales)")
plt.title("Distribuição das vendas por plataforma (boxplot)")
ax = plt.gca()
ax.yaxis.set_major_locator(MultipleLocator(3))
plt.tight_layout()
plt.show()

medias = dados.groupby('platform')['total_sales'].mean().sort_values(ascending=False)
print("Médias de vendas por plataforma:\n")
print(medias.head(20))

print("\nMedianas de vendas por plataforma:\n")
print(medianas.head(20))
```

```python
plataforma_base = "PS3"
comparar_com    = "X360"
ano_ini, ano_fim = 2012, 2015
filtro = ((df_games["year_of_release"] >= ano_ini) & (df_games["year_of_release"] <= ano_fim) & (df_games["platform"].isin([plataforma_base, comparar_com])))
dados = df_games.loc[filtro, ["name", "platform", "total_sales"]].dropna()

vendas_por_jogo_plataforma = dados.groupby(["name", "platform"], as_index=False)["total_sales"].sum()

comparativo_mesmos_jogos = vendas_por_jogo_plataforma.pivot(index="name", columns="platform", values="total_sales")
comparativo_mesmos_jogos = comparativo_mesmos_jogos.dropna(subset=[plataforma_base, comparar_com])

comparativo_mesmos_jogos["soma"] = comparativo_mesmos_jogos[plataforma_base] + comparativo_mesmos_jogos[comparar_com]
comparativo_mesmos_jogos = comparativo_mesmos_jogos.sort_values("soma", ascending=False)

col_dif = f"dif_{plataforma_base}_menos_{comparar_com}"
comparativo_mesmos_jogos[col_dif] = comparativo_mesmos_jogos[plataforma_base] - comparativo_mesmos_jogos[comparar_com]

def vencedor(row):
    if row[plataforma_base] > row[comparar_com]:
        return plataforma_base
    elif row[plataforma_base] < row[comparar_com]:
        return comparar_com
    else:
        return "empate"

comparativo_mesmos_jogos["vence"] = comparativo_mesmos_jogos.apply(vencedor, axis=1)

print(comparativo_mesmos_jogos[[plataforma_base, comparar_com, col_dif, "vence"]].head(15))

total_comp = len(comparativo_mesmos_jogos)
v_base  = (comparativo_mesmos_jogos["vence"] == plataforma_base).sum()
v_comp  = (comparativo_mesmos_jogos["vence"] == comparar_com).sum()
empates = (comparativo_mesmos_jogos["vence"] == "empate").sum()

print("\nResumo:")
print("Jogos comparados:", total_comp)
print(f"{plataforma_base} venceu em:", v_base)
print(f"{comparar_com} venceu em:", v_comp)
print("Empates:", empates)
```

Observando a distribuição por gênero no período 2012–2015, os gêneros mais lucrativos são aqueles no topo da coluna de vendas_totais e com alta participação na porcentagem. Em geral isso acontece porque há muitos lançamentos e/ou alguns grandes sucessos. 
Para medir o retorno normal por jogo, a maneira de calcular mais próxima da realidade é a mediana_venda_por_jogo, pois sofre menos com outliers. Se a média ficar muito acima da mediana, significa que poucos títulos grandes estão puxando a média para cima, deixando o gênero mais arriscado (dependente de sucessos). 
Em termos gerais, gêneros com vendas altas costumam combinar uma base grande de jogadores com franquias fortes. Os gêneros com vendas baixas aparecem com vendas totais menores e, muitas vezes, mediana modesta.

```python
df_model = df_games[(df_games['year_of_release'] >= 2012) & (df_games['year_of_release'] <= 2015)].copy()
df_model['rating'] = df_model['rating'].fillna('Unknown')

regioes = [('AN', 'na_sales'),('UE', 'eu_sales'),('JP', 'jp_sales'),]

top_plataformas_por_regiao = {}

for rotulo_regiao, coluna_vendas_regiao in regioes:
    print("\nRegião:", rotulo_regiao, "| Coluna de vendas:", coluna_vendas_regiao)
    vendas_totais_regiao = df_model[coluna_vendas_regiao].sum()
    vendas_por_plataforma = df_model.groupby('platform')[coluna_vendas_regiao].sum()
    vendas_por_plataforma = vendas_por_plataforma.sort_values(ascending=False)
    
    top5_plataformas = vendas_por_plataforma.head(5)
    top_plataformas_por_regiao[rotulo_regiao] = list(top5_plataformas.index)
    print("\nTop 5 plataformas (vendas e quota %):")
    for plataforma_nome in top5_plataformas.index:
        vendas_plataforma = float(top5_plataformas.loc[plataforma_nome])
        if vendas_totais_regiao == 0:
            quota_percentual = 0.0
        else:
            quota_percentual = (vendas_plataforma / vendas_totais_regiao) * 100.0

        print(plataforma_nome, "| vendas =", round(vendas_plataforma, 2), "| quota =", round(quota_percentual, 1), "%")

    vendas_por_genero = df_model.groupby('genre')[coluna_vendas_regiao].sum()
    vendas_por_genero = vendas_por_genero.sort_values(ascending=False)
    top5_generos = vendas_por_genero.head(5)
    print("\nTop 5 gêneros (vendas):")
    for genero_nome in top5_generos.index:
        vendas_genero = float(top5_generos.loc[genero_nome])
        print(genero_nome, "| vendas =", round(vendas_genero, 2))

    vendas_por_rating = df_model.groupby('rating')[coluna_vendas_regiao].sum().sort_values(ascending=False)
    media_por_rating = df_model.groupby('rating')[coluna_vendas_regiao].mean()
    print("\nESRB — vendas totais e média por jogo:")
    for rating_nome in vendas_por_rating.index:
        total_rating = float(vendas_por_rating.loc[rating_nome])
        media_rating = float(media_por_rating.loc[rating_nome])
        print("Rating:", rating_nome, "| total =", round(total_rating, 2), "| média =", round(media_rating, 2))

plataformas_destaque = []
for rotulo_regiao in ['AN', 'UE', 'JP']:
    nomes_plataformas = top_plataformas_por_regiao.get(rotulo_regiao, [])
    for nome in nomes_plataformas:
        if nome not in plataformas_destaque:
            plataformas_destaque.append(nome)

total_AN = float(df_model['na_sales'].sum())
total_UE = float(df_model['eu_sales'].sum())
total_JP = float(df_model['jp_sales'].sum())

print("\n===============================================")
print("Quotas por região — Plataformas em destaque (2012–2015):")

for plataforma_nome in plataformas_destaque:
    vendas_AN_plat = float(df_model.loc[df_model['platform'] == plataforma_nome, 'na_sales'].sum())
    vendas_UE_plat = float(df_model.loc[df_model['platform'] == plataforma_nome, 'eu_sales'].sum())
    vendas_JP_plat = float(df_model.loc[df_model['platform'] == plataforma_nome, 'jp_sales'].sum())

    if total_AN == 0.0:
        quota_AN_percent = 0.0
    else:
        quota_AN_percent = (vendas_AN_plat / total_AN) * 100.0

    if total_UE == 0.0:
        quota_UE_percent = 0.0
    else:
        quota_UE_percent = (vendas_UE_plat / total_UE) * 100.0

    if total_JP == 0.0:
        quota_JP_percent = 0.0
    else:
        quota_JP_percent = (vendas_JP_plat / total_JP) * 100.0

    print(plataforma_nome,
          "| AN =", round(quota_AN_percent, 1), "%",
          "| UE =", round(quota_UE_percent, 1), "%",
          "| JP =", round(quota_JP_percent, 1), "%")
```

Considerando 2012–2015, o perfil regional mostra diferenças claras. 
Na América do Norte (NA), há forte participação de X360 e PS3, com PS4 e XOne já relevantes, enquanto o 3DS mantém presença moderada. 

Na Europa (UE), PS3/PS4 lideram com folga, X360/XOne ficam atrás e o 3DS segue consistente. 

No Japão (JP), o 3DS domina com folga, o PS3 ainda é importante, o PS4 aparece menor, e X360/XOne são quase inexistentes.

Nos gêneros, NA e UE têm maior peso de Action, Shooter e Sports, enquanto no JP há mais espaço para Role-Playing (RPG) e Action, com Shooter menos dominante que no Ocidente. A preferência cultural e o peso de franquias locais explicam bem esse desvio (como o protagonismo de RPGs no Japão).

```python
from scipy import stats as st

df_teste = df_games.copy()
df_teste['user_score'] = pd.to_numeric(df_teste['user_score'], errors='coerce')

df_teste = df_teste[
    (df_teste['year_of_release'] >= 2012) & (df_teste['year_of_release'] <= 2015)].dropna(subset=['user_score'])


alpha = 0.05 #definir nível de significância
# HIPÓTESE A: XOne vs PC (médias iguais?)
scores_xone = df_teste.loc[df_teste['platform'] == 'XOne', 'user_score']
scores_pc   = df_teste.loc[df_teste['platform'] == 'PC',   'user_score']

media_xone = scores_xone.mean()
media_pc   = scores_pc.mean()

t_xone_pc, p_xone_pc = st.ttest_ind(scores_xone, scores_pc, equal_var=False, nan_policy='omit')

print("HIPÓTESE A — XOne vs PC")
print("n_XOne =", len(scores_xone), "| n_PC =", len(scores_pc))
print("média_XOne =", round(float(media_xone), 3), "| média_PC =", round(float(media_pc), 3))
print("t =", round(float(t_xone_pc), 3), "| p =", round(float(p_xone_pc), 4))

if p_xone_pc < alpha:
    print("Conclusão: rejeitar H0 (médias diferentes ao nível de", alpha, ")")
else:
    print("Conclusão: não rejeitar H0 (sem evidência de diferença ao nível de", alpha, ")")

print("-" * 60)

# HIPÓTESE B: Action vs Sports (médias diferentes?)
scores_action = df_teste.loc[df_teste['genre'] == 'Action',  'user_score']
scores_sports = df_teste.loc[df_teste['genre'] == 'Sports',  'user_score']

media_action  = scores_action.mean()
media_sports  = scores_sports.mean()

t_act_spt, p_act_spt = st.ttest_ind(scores_action, scores_sports, equal_var=False, nan_policy='omit')

print("HIPÓTESE B — Action vs Sports")
print("n_Action =", len(scores_action), "| n_Sports =", len(scores_sports))
print("média_Action =", round(float(media_action), 3), "| média_Sports =", round(float(media_sports), 3))
print("t =", round(float(t_act_spt), 3), "| p =", round(float(p_act_spt), 4))

if p_act_spt < alpha:
    print("Conclusão: rejeitar H0 (médias diferentes ao nível de", alpha, ")")
else:
    print("Conclusão: não rejeitar H0 (sem evidência de diferença ao nível de", alpha, ")")

```

Usei comparação de médias com teste bicaudal.

Hipótese A (XOne vs PC)
H0: as médias de user_score são iguais.
H1: as médias são diferentes.

Hipótese B (Action vs Sports)
H0: as médias de user_score são iguais.
H1: as médias são diferentes.

Nível de significância (alpha)
Escolhi α = 0,05 por ser o padrão em testes estatísticos: equilibra bem risco de falso positivo (rejeitar H0 quando é verdadeira) com praticidade. Se o seu uso exigir mais rigor, você pode repetir com α = 0,01.

Por que Welch t-test (equal_var=False)
É o teste t independente que não assume variâncias iguais entre os grupos. É uma escolha segura e simples para dados reais onde as variâncias e tamanhos de amostra podem ser diferentes.
